```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Importing the data
in_dt = pd.read_csv('C:/Users/jatin/Desktop/wine-data.csv')
in_dt
```

```python
# Check data for missing values
in_dt.info()
```

```python
# Check data for any other inconsistencies
in_dt.describe()
```

```python
# Removing the column "class"
in_dt1 = in_dt.iloc[:,1:14]
in_dt2 = in_dt.iloc[:,1:14]
in_dt3 = in_dt.iloc[:,1:14]
in_dt1.head()
```

```python
in_dt1.shape
```

```python
# Standardize the data so that all the varibales are in scale

from sklearn import preprocessing

stzd_dt = preprocessing.StandardScaler().fit_transform(in_dt1)
stzd_dt = pd.DataFrame(stzd_dt, columns=in_dt1.columns)
stzd_dt.head()
```

```python
stzd_dt.shape
```

```python
# Applying PCA

from sklearn.decomposition import PCA

pca_dt = PCA().fit(stzd_dt)

# Proportion of Variance

pca_dt.explained_variance_ratio_
```

```python
# Cumulative proportion of variance
a = np.cumsum(pca_dt.explained_variance_ratio_)
              
print("Cumulative proportion of variance:", a)
```

```python
# component loadings/Weights

loadings = pca_dt.components_
features = pca_dt.n_features_
pca_list = ["PC"+str(i) for i in list(range(1, features+1))]
loadings_dt = pd.DataFrame.from_dict(dict(zip(pca_list, loadings)))
loadings_dt['Attributes'] = stzd_dt.columns.values
loadings_dt = loadings_dt.set_index('Attributes')
loadings_dt
```

```python
# Eigen Values of Principal components

b = pca_dt.explained_variance_

print("Eigen Values of Principle components:", b)
```

```python
# Scree Plot

from bioinfokit.visuz import cluster

cluster.screeplot(obj=[pca_list, pca_dt.explained_variance_ratio_], show=True, dim=(20,10), axlabelfontsize = 18)
```

```python
# Plot Correlataion matrix for first 3 Principal Components

import seaborn as sns

loadings_dt1 = loadings_dt.iloc[:,0:3]
plt.figure(figsize=(8,6))
cor_plt = sns.heatmap(loadings_dt1, annot=True, cmap='coolwarm')
plt.show()
```

```python
# 2D Biplot for Attributes

target = in_dt['Class'].to_numpy()

pca_dt1 = PCA().fit_transform(stzd_dt)

cluster.biplot(cscore=pca_dt1, loadings=pca_dt.components_, labels=in_dt1.columns.values, var1=round(pca_dt.explained_variance_ratio_[0]*100, 2),
    var2=round(pca_dt.explained_variance_ratio_[1]*100, 2), show=True, dim=(20,10), axlabelfontsize = 18,  colorlist=target)

```

## K-Means Clustering using all chemical measures

```python
# K-means clustering
from sklearn.cluster import KMeans
fit = KMeans(n_clusters=3, max_iter = 10, random_state=0).fit(stzd_dt)

# Determine number of clusters
Cluster_Variability = []
for i in range(1,6): 
    kmeans =  KMeans(n_clusters=i).fit(stzd_dt)
    Cluster_Variability.append(kmeans.inertia_)
#Plot the elbow graph
plt.figure(figsize=(20,10))
plt.plot(range(1,6),Cluster_Variability,marker='o')
plt.title('The Elbow Method Graph')
plt.xticks(range(1,6))
plt.xlabel('Number of clusters')
plt.ylabel('Within groups sum of squares')
plt.show()
```

```python
Kmeans_cluster = KMeans(n_clusters=3, max_iter = 10, random_state=0).fit(stzd_dt)
in_dt2["K-means_Cluster_ID"]=Kmeans_cluster.labels_
in_dt2.head()
```

```python
in_dt2.groupby("K-means_Cluster_ID").mean()
```

```python
# Number of Wines in each Cluster
in_dt2['K-means_Cluster_ID'].value_counts()
```

## K-Means Clustering using two most significant PC scores

```python
# Clustering using two most significant PC scores

#Define two components  
pca=PCA(n_components=2)

# Applying PCA
pca_dt2 = pca.fit_transform(stzd_dt) 
pc_dt = pd.DataFrame(data=pca_dt2,columns=['PC1','PC2'])
pc_dt
```

```python
# K-means clustering
from sklearn.cluster import KMeans
fit = KMeans(n_clusters=3, max_iter = 10, random_state=0).fit(pc_dt)

# Determine number of clusters
Cluster_Variability = []
for i in range(1,6): 
    kmeans =  KMeans(n_clusters=i).fit(stzd_dt)
    Cluster_Variability.append(kmeans.inertia_)
#Plot the elbow graph
plt.figure(figsize=(20,10))
plt.plot(range(1,6),Cluster_Variability,marker='o')
plt.title('The Elbow Method Graph')
plt.xticks(range(1,6))
plt.xlabel('Number of clusters')
plt.ylabel('Within groups sum of squares')
plt.show()
```

```python
Kmeans_cluster1 = KMeans(n_clusters=3, max_iter = 10, random_state=0).fit(pc_dt)
in_dt3["K-means_PCA_Cluster"]=Kmeans_cluster1.labels_
in_dt3.head()
```

```python
in_dt3.groupby("K-means_PCA_Cluster").mean()
```

```python
# Number of Wines in each Cluster
in_dt3['K-means_PCA_Cluster'].value_counts()
```

```python
# Plotting clusters using 2 PC scores

kmeans = KMeans(n_clusters= 3)
 
y_kmeans = kmeans.fit_predict(pca_dt2)

plt.figure(figsize=(20,10))
plt.scatter(pca_dt2[y_kmeans==0, 0], pca_dt2[y_kmeans==0, 1], s=100, c='red', label ='Cluster 0')
plt.scatter(pca_dt2[y_kmeans==1, 0], pca_dt2[y_kmeans==1, 1], s=100, c='blue', label ='Cluster 1')
plt.scatter(pca_dt2[y_kmeans==2, 0], pca_dt2[y_kmeans==2, 1], s=100, c='green', label ='Cluster 2')
plt.legend()
plt.title('K-Means Clustering using two most significant PC scores')
plt.show()
```
